# Performance Analysis

```python
df = pd.read_csv("./data/results.csv")
df
```

```python
sns.pairplot(df)
plt.savefig('./plots_perf/data_pair.png')
plt.show()
```

```python
plt.figure(figsize=(40, 20))
plt.subplot(1,2,1)
sns.boxplot(x = 'm1_action', y = 'adj_revenue_m1', data = df).set(
    xlabel='Action Type (Previous Model)', 
    ylabel='Adjusted Revenue'
)
plt.subplot(1,2,2)
sns.boxplot(x = 'm2_action', y = 'adj_revenue_m2', data = df).set(
    xlabel='Action Type (Proposed Model)', 
    ylabel='Adjusted Revenue'
)
plt.savefig('./plots_perf/action_dist.png')
plt.show()
```

```python
df
```

```python
results1 = df[['day', 'transaction', 'm1_action', 'm1_action_cost','adj_revenue_m1'] ]
results1
```

```python
results1['Model'] = 'M1'
results1
```

```python
results1 = results1.rename(columns={"m1_action": "action", "m1_action_cost": "action_cost", "adj_revenue_m1": "adj_revenue"})
```

```python
results2 = df[['day', 'transaction','m2_action', 'm2_action_cost','adj_revenue_m2']]
results2
```

```python
results2['Model'] = 'M2'
results2
```

```python
results2 = results2.rename(columns={"m2_action": "action", "m2_action_cost": "action_cost", "adj_revenue_m2": "adj_revenue"})
results2
```

```python
results1
```

```python
results2
```

```python
results1['cum_revenue'] = results1['adj_revenue'].cumsum()
results1
```

```python
results2['cum_revenue'] = results2['adj_revenue'].cumsum()
results2
```

```python
results = results1.append(results2, ignore_index=True)
results
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x = results['action'],
            y = results['adj_revenue'],
            hue = results['Model']).set(
        xlabel='Action Type', 
        ylabel='Adjusted Revenue'
)
# plt.legend(labels=["Previous Model","Proposed Model"], fontsize = 20)
plt.savefig('./plots_perf/action_dist_gp.png')
plt.show()
```

## Revenue by Transaction Amount Comparision

```python
plt.figure(figsize=(20, 10))
# sns.lineplot(data=results, x="transaction", y="adj_revenue", hue="model")
sns.lineplot(data=results, x="transaction", y="adj_revenue", hue="Model").set(
        xlabel='Transaction Amount', 
        ylabel='Adjusted Revenue'
)
# plt.legend(labels=["Previous Model","Proposed Model"], fontsize = 20)
plt.savefig('./plots_perf/revenue_by_txn.png')
plt.show()
```

## Cumulative Revenue Comparision

```python
results
```

```python
plt.figure(figsize=(30, 20))
sns.lineplot(data=results, x="day", y="cum_revenue", hue="Model")
plt.xlabel("Day", fontsize = 20)
plt.ylabel("Cumulative Revenue", fontsize = 20)
plt.legend(fontsize = 20)
plt.savefig('./plots_perf/cumulative_revenue.png')
plt.show()
```
